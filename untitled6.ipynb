```python
# Big Data Analytics with LLM - OpenAI ChatGPT Integration

# Step 1: Setup and Installation
!pip install openai pandas matplotlib seaborn scikit-learn transformers nltk

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
import openai
import json
import time
from datetime import datetime
import re

# Download required NLTK data
nltk.download('vader_lexicon')
nltk.download('punkt')

# Step 2: Configure OpenAI API (you'll need to provide your API key)
# Replace 'your-api-key' with your actual OpenAI API key
openai.api_key = ""

# Step 3: Load and Explore Sample Dataset
# For this example, we'll use a sample customer feedback dataset

# Generate sample data for demonstration
def generate_sample_data(n=1000):
    # Create date range
    dates = pd.date_range(start='2023-01-01', periods=n)

    # Create sample customer feedbacks
    feedbacks = [
        "I love this product, it's amazing!",
        "The service was terrible, I'm disappointed.",
        "Average experience, nothing special.",
        "Great customer support, very helpful.",
        "The product quality is poor.",
        "Excellent value for money.",
        "Shipping was delayed, not happy.",
        "The app keeps crashing, please fix it.",
        "I would definitely recommend this to others.",
        "The new features are very useful."
    ]

    # Create sample data
    data = {
        'date': dates,
        'customer_id': np.random.randint(1000, 9999, size=n),
        'product_category': np.random.choice(['Electronics', 'Clothing', 'Home', 'Books', 'Food'], size=n),
        'rating': np.random.randint(1, 6, size=n),
        'feedback': np.random.choice(feedbacks, size=n),
        'purchase_amount': np.random.uniform(10, 500, size=n).round(2)
    }

    return pd.DataFrame(data)

# Generate and display sample data
df = generate_sample_data()
print("Sample dataset shape:", df.shape)
df.head()

# Step 4: Data Preprocessing and Exploration

# Basic statistics
print("\nBasic Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

# Distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df, palette='viridis')
plt.title('Distribution of Customer Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.show()

# Sales by product category
plt.figure(figsize=(12, 6))
category_sales = df.groupby('product_category')['purchase_amount'].sum().sort_values(ascending=False)
sns.barplot(x=category_sales.index, y=category_sales.values, palette='viridis')
plt.title('Total Sales by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.show()

# Time series analysis of sales
df['month'] = df['date'].dt.strftime('%Y-%m')
monthly_sales = df.groupby('month')['purchase_amount'].sum().reset_index()
monthly_sales['month'] = pd.to_datetime(monthly_sales['month'])
monthly_sales = monthly_sales.sort_values('month')

plt.figure(figsize=(14, 6))
plt.plot(monthly_sales['month'], monthly_sales['purchase_amount'], marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales Amount')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# Step 5: Sentiment Analysis using NLTK
sia = SentimentIntensityAnalyzer()

def analyze_sentiment(text):
    sentiment = sia.polarity_scores(text)
    if sentiment['compound'] >= 0.05:
        return 'Positive'
    elif sentiment['compound'] <= -0.05:
        return 'Negative'
    else:
        return 'Neutral'

df['sentiment'] = df['feedback'].apply(analyze_sentiment)
df['sentiment_score'] = df['feedback'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Plot sentiment distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='sentiment', data=df, palette='viridis')
plt.title('Distribution of Sentiment in Customer Feedback')
plt.xlabel('Sentiment')
plt.ylabel('Count')
plt.show()

# Relationship between sentiment and rating
plt.figure(figsize=(10, 6))
sns.boxplot(x='rating', y='sentiment_score', data=df, palette='viridis')
plt.title('Relationship Between Rating and Sentiment Score')
plt.xlabel('Rating')
plt.ylabel('Sentiment Score')
plt.grid(True, alpha=0.3)
plt.show()

# Step 6: Advanced Analytics with OpenAI ChatGPT Integration

def analyze_feedback_with_gpt(feedback, max_retries=3, delay=2):
    """
    Analyze customer feedback using OpenAI GPT

    Parameters:
    feedback (str): Customer feedback text
    max_retries (int): Maximum number of retries on API error
    delay (int): Delay between retries in seconds

    Returns:
    dict: Analysis results
    """
    prompt = f"""
    Analyze the following customer feedback and provide:
    1. A brief summary (max 10 words)
    2. The key issues or praises mentioned
    3. Suggested actions for the company
    4. The emotional tone (frustrated, satisfied, neutral, etc.)

    Format the response as a JSON with keys: 'summary', 'key_points', 'suggested_actions', 'emotional_tone'

    Customer Feedback: "{feedback}"
    """

    for attempt in range(max_retries):
        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",
                messages=[
                    {"role": "system", "content": "You are a customer feedback analysis assistant. Provide insights in JSON format only."},
                    {"role": "user", "content": prompt}
                ],
                temperature=0.3,
                max_tokens=300
            )

            result_text = response.choices[0].message.content

            # Extract the JSON part
            json_match = re.search(r'({.*})', result_text, re.DOTALL)
            if json_match:
                result_text = json_match.group(1)

            return json.loads(result_text)

        except (openai.error.RateLimitError, openai.error.APIError, json.JSONDecodeError) as e:
            if attempt == max_retries - 1:
                print(f"Failed after {max_retries} attempts: {e}")
                return {
                    'summary': 'Analysis failed',
                    'key_points': [],
                    'suggested_actions': [],
                    'emotional_tone': 'unknown'
                }
            print(f"Attempt {attempt+1} failed: {e}. Retrying in {delay} seconds...")
            time.sleep(delay)

# To conserve API calls, we'll analyze just a sample of the feedback
sample_size = 5  # In a real scenario, you might want to process more
sample_feedbacks = df.sample(sample_size)['feedback'].tolist()

analysis_results = []
for feedback in sample_feedbacks:
    print(f"Analyzing: {feedback}")
    # Uncomment the following line to actually call the API
    # result = analyze_feedback_with_gpt(feedback)
    # For demonstration purposes, we'll use mock results
    result = {
        'summary': 'Mock summary of feedback',
        'key_points': ['Point 1', 'Point 2'],
        'suggested_actions': ['Action 1', 'Action 2'],
        'emotional_tone': np.random.choice(['positive', 'negative', 'neutral'])
    }
    analysis_results.append(result)
    print(f"Result: {result}")

# Create a DataFrame from the analysis results
gpt_analysis_df = pd.DataFrame(analysis_results)
gpt_analysis_df['feedback'] = sample_feedbacks

# Display the GPT analysis results
gpt_analysis_df

# Step 7: Customer Segmentation using K-means Clustering

# Select and prepare numerical features
features = df[['rating', 'sentiment_score', 'purchase_amount']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Determine optimal number of clusters using the Elbow Method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True, alpha=0.3)
plt.show()

# Based on the elbow method, choose an appropriate number of clusters
optimal_k = 4  # This should be determined from the plot above
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
df['cluster'] = kmeans.fit_predict(scaled_features)

# Visualize the clusters using PCA for dimensionality reduction
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features)
df['pca1'] = pca_result[:, 0]
df['pca2'] = pca_result[:, 1]

plt.figure(figsize=(12, 8))
sns.scatterplot(x='pca1', y='pca2', hue='cluster', data=df, palette='viridis', s=100, alpha=0.7)
plt.title('Customer Segments Visualization using PCA')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Cluster')
plt.grid(True, alpha=0.3)
plt.show()

# Analyze cluster characteristics
cluster_analysis = df.groupby('cluster').agg({
    'rating': 'mean',
    'sentiment_score': 'mean',
    'purchase_amount': 'mean',
    'customer_id': 'count'
}).rename(columns={'customer_id': 'count'}).reset_index()

# Visualize cluster characteristics
plt.figure(figsize=(14, 10))

# Plot 1: Average Rating by Cluster
plt.subplot(2, 2, 1)
sns.barplot(x='cluster', y='rating', data=cluster_analysis, palette='viridis')
plt.title('Average Rating by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Rating')

# Plot 2: Average Sentiment Score by Cluster
plt.subplot(2, 2, 2)
sns.barplot(x='cluster', y='sentiment_score', data=cluster_analysis, palette='viridis')
plt.title('Average Sentiment Score by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Sentiment Score')

# Plot 3: Average Purchase Amount by Cluster
plt.subplot(2, 2, 3)
sns.barplot(x='cluster', y='purchase_amount', data=cluster_analysis, palette='viridis')
plt.title('Average Purchase Amount by Cluster')
plt.xlabel('Cluster')
plt.ylabel('Average Purchase Amount')

# Plot 4: Cluster Size
plt.subplot(2, 2, 4)
sns.barplot(x='cluster', y='count', data=cluster_analysis, palette='viridis')
plt.title('Number of Customers in Each Cluster')
plt.xlabel('Cluster')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Step 8: Integrate GPT insights with clusters (conceptual)
# In a real application, you would process all feedbacks through GPT
# For demonstration, we'll simulate combining the cluster data with GPT insights

# Step 9: Create an interactive dashboard (conceptual)
# In a real application, you would use libraries like Plotly Dash or Streamlit
# For demonstration, we'll create a summary visualization

# Summary of key metrics by product category and sentiment
category_sentiment = df.groupby(['product_category', 'sentiment']).agg({
    'rating': 'mean',
    'purchase_amount': 'mean',
    'customer_id': 'count'
}).rename(columns={'customer_id': 'count'}).reset_index()

plt.figure(figsize=(16, 12))

# Plot 1: Average Rating by Category and Sentiment
plt.subplot(2, 2, 1)
sns.barplot(x='product_category', y='rating', hue='sentiment', data=category_sentiment, palette='viridis')
plt.title('Average Rating by Category and Sentiment')
plt.xlabel('Product Category')
plt.ylabel('Average Rating')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')

# Plot 2: Average Purchase Amount by Category and Sentiment
plt.subplot(2, 2, 2)
sns.barplot(x='product_category', y='purchase_amount', hue='sentiment', data=category_sentiment, palette='viridis')
plt.title('Average Purchase Amount by Category and Sentiment')
plt.xlabel('Product Category')
plt.ylabel('Average Purchase Amount')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')

# Plot 3: Customer Count by Category and Sentiment
plt.subplot(2, 2, 3)
sns.barplot(x='product_category', y='count', hue='sentiment', data=category_sentiment, palette='viridis')
plt.title('Customer Count by Category and Sentiment')
plt.xlabel('Product Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Sentiment')

# Plot 4: Distribution of Clusters by Product Category
plt.subplot(2, 2, 4)
cluster_category = pd.crosstab(df['cluster'], df['product_category'])
cluster_category.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Distribution of Clusters by Product Category')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.legend(title='Product Category')

plt.tight_layout()
plt.show()

# Step 10: GPT-Based Recommendations (conceptual)
# In a real application, you would generate recommendations using GPT based on the analysis
# For demonstration, we'll use a sample prompt

recommendation_prompt = """
Based on the customer feedback analysis and segmentation, provide strategic recommendations to improve:
1. Customer satisfaction for each identified cluster
2. Product improvements for categories with lower ratings
3. Marketing strategies tailored to different customer segments

Provide 1-2 specific actionable recommendations for each point.
"""

# Uncomment to get actual recommendations from GPT
# recommendations = openai.ChatCompletion.create(
#     model="gpt-3.5-turbo",
#     messages=[
#         {"role": "system", "content": "You are a business strategy consultant providing data-driven recommendations."},
#         {"role": "user", "content": recommendation_prompt}
#     ],
#     temperature=0.7,
#     max_tokens=500
# )
# print(recommendations.choices[0].message.content)

# For demonstration, we'll show a sample recommendation output
print("\nSample GPT-Generated Strategic Recommendations:")
print("""
# Strategic Recommendations Based on Data Analysis

## Customer Satisfaction Improvement by Cluster
1. **Cluster 0 (High spenders with moderate satisfaction)**: Implement a premium customer service line and personalized follow-ups after purchases to increase loyalty.
2. **Cluster 1 (Low-rated, negative sentiment)**: Establish a dedicated customer recovery program with special offers and direct communication channels.
3. **Cluster 2 (Average ratings and sentiment)**: Create an enhanced loyalty program with tiered benefits to encourage more engagement and higher spending.
4. **Cluster 3 (High satisfaction, lower spending)**: Develop cross-selling strategies and bundle offers to increase purchase amounts while maintaining high satisfaction.

## Product Improvement Recommendations
1. **Electronics category**: Focus on improving product quality and reliability based on the negative feedback patterns identified.
2. **Clothing category**: Enhance size consistency and material quality as these were common issues in customer feedback.

## Targeted Marketing Strategies
1. **For positive sentiment clusters**: Leverage customer testimonials and implement referral programs to attract similar customers.
2. **For negative sentiment clusters**: Develop educational content addressing common pain points and showcase product improvements.

Implementing these recommendations could potentially increase overall customer satisfaction by 15% and boost sales in underperforming categories by 20% within 6 months.
""")

# Step 11: Export Results (optional)
# df.to_csv('customer_analysis_results.csv', index=False)
# cluster_analysis.to_csv('cluster_characteristics.csv', index=False)

print("\nAnalysis complete! The notebook has demonstrated how to integrate LLM capabilities with traditional data analysis techniques.")
```
